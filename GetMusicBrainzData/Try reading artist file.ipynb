```python
artists = []
```

```python
for i in range(1, len(lines)):
    try:
        artists.append(lines[i].split('\t')[2])
    except IndexError : 
        print(lines[i])
```

```python
english_artists = []
```

```python
for i in range(len(artists)):
    try:
        if detect(artists[i]) == 'en':
            english_artists.append(artists[i])
    except :
        print(artists[i])
```

```python
english_artists
```

```python
len(english_artists)
```

```python
import pickle
```

```python
with open('en_artists_from_musicbrainz', 'wb') as fp:
    pickle.dump(english_artists, fp)
```
